```python
import numpy as np
import matplotlib.pyplot as plt

# Define input parameters
num_simulations = 1000
num_years = 10
initial_price = 100
mu = 0.05
sigma = 0.2

# Calculate drift and volatility
dt = 1/365
drift = (mu - 0.5 * sigma**2) * dt
volatility = sigma * np.sqrt(dt)

# Generate random price paths
price_paths = np.zeros((num_simulations, num_years*365))
price_paths[:,0] = initial_price

for i in range(num_years*365-1):
    random_numbers = np.random.normal(0, 1, num_simulations)
    price_paths[:,i+1] = price_paths[:,i] * np.exp(drift + volatility * random_numbers)

# Plot results
plt.plot(price_paths.T)
plt.xlabel('Days')
plt.ylabel('Price')
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.95, top=0.95, wspace=0.2, hspace=0.5)
plt.show()
```
